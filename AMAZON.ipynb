```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt

import pandas as pd
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
swords = set(stopwords.words('english'))
swords
```

```python
amazon = pd.read_excel('D:/NAYANA/Term 1/EDA/Files/Amazon.xlsx')
amazon.head(2)
```

```python
text = amazon[['reviews.text']]
print(text['reviews.text'][0])
text.head()
```

```python
# Converting everything into lower case

text['clean_text'] = text['reviews.text'].str.lower().str.replace('[^a-z ]', '')
text.head()
```

```python
# Removing the stopwords

stop = stopwords.words('english')
stop.extend(['amazon'])

def sw(text) :
    text = [word for word in text.split() if word not in stop]
    return ' '.join(text)

# Applying the sw on clean_text columns

text['text_without_stop'] = text['clean_text'].apply(sw)
print(text['clean_text'][0])
print(text['text_without_stop'][0])
```

```python
# Starting the count vector

vec = CountVectorizer()
vec.fit(text['text_without_stop'])
```

```python
# DTM matrix

x = vec.fit_transform(text['text_without_stop'])
x
```

```python
# Conc=verting this DTM into dataframes

DTM = pd.DataFrame(x.toarray(), columns = vec.get_feature_names())
DTM.head()
```

```python
# Calculating the word freq

word_freq = DTM.sum()
word_freq.sort_values(ascending = False).head()
```
